### Exercise 9: Merging two DataFrames ###
Below are two DataFrames of albums. Merge them into one DataFrame, that has two coloumns: "name" and "year".

```python
floyd_albums = pd.DataFrame({
    "album": ["Atom Heart Mother", "The Dark Side of the Moon", "Wish You Were Here", "The Wall"],
    "year": [1970,1973,1975,1979]
})

crimson_albums = pd.DataFrame({
    "name": ["In The Court of the Crimson King","Lizard", "Larks' Tongues in Aspic", "Red", "Discipline"], 
    "released": [1969, 1970, 1973, 1974, 1981]
})

### YOUR CODE HERE ###
floyd_albums = floyd_albums.rename(columns={"album": "name"})
crimson_albums = crimson_albums.rename(columns={"released": "year"})

albums = floyd_albums.merge(crimson_albums, "outer").sort_values(by="year")
albums
```

```python
X = ['Group A','Group B','Group C','Group D']
girls = [10,20,20,40]
boys = [20,30,25,30]
  
X_axis = np.arange(len(X))
print(X_axis)
plt.bar(X_axis - 0.2, girls, 0.4, label = 'Girls', hatch='/')
plt.bar(X_axis + 0.2, boys, 0.4, label = 'Boys')

y_errormin =[1, 5, 9, 1]
y_errormax =[2, 4, 6,4]
y_error = [y_errormin, y_errormax]
#y_error = 5
plt.errorbar(X_axis - 0.2, girls, yerr = y_error, fmt ='o', c = 'k')

plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
